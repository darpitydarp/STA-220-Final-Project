Merging Datasets

```python
import pandas as pd

df = pd.read_csv("temperature.csv")
counts = pd.read_csv("ecocounter_bike_traffic_data.csv")
```

```python
temperature = df[["DATE", "LATITUDE", "LONGITUDE", "TOBS"]]
temperature = temperature.rename(columns={'DATE': 'Date', 'LATITUDE': 'latitude', 'LONGITUDE': 'longitude', 'TOBS': 'temperature'})
temperature
```

```python
data = pd.merge(counts, temperature, on='Date', how='left')
data
```

```python
data.to_csv("merged_data.csv", index=False)
```
